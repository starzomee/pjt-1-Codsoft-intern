```python
# let's find the index in "Age" column where value is missing 
missing_indices = full_data[full_data['Age'].isnull()].index
missing_indices
```

```python
# now calculate mean of age for each class
mean_age_by_class = full_data.groupby('Pclass')['Age'].mean()
mean_age_by_class
```

```python
# now we fill missing values in "Age" column by mean 
# impute function 
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37 
        elif Pclass == 2:
            return 29

        else:
            return 24 

    else:
        return Age 


full_data['Age'] = full_data[['Age' , 'Pclass']].apply(impute_age , axis = 1)   
```

```python
# let's check it now 
full_data["Age"].isnull().sum()
```

```python
# lets check it in heatmap 
sns.heatmap(full_data.isnull() , yticklabels = False , cbar = False , cmap = "tab20c_r")
plt.title("Check for missing values")
plt.show()
```

```python
# let's find the index in "Cabin" column where value is missing 
missing_indices_Cabin = full_data[full_data['Cabin'].isnull()].index
missing_indices_Cabin
```

```python
# now we see we have filled the missing values of "Age" column 
# now we deal with "Cabin" as it is not as such important and also it has almost 70% of missing values so we drop it 
full_data.drop("Cabin" , axis = 1 , inplace = True)
```

```python
gender_conversion = {"male" : 0 , "female" : 1}
data = [full_data]
for dataset in data:
    dataset['Sex'] = dataset['Sex'].map(gender_conversion)
```

```python
# Now let's do the same with "Embarked"
embarked_conversion = {'C':0  , 'Q':1 , 'S':2}
data = [full_data]
for dataset in data:
    dataset['Embarked'] = dataset['Embarked'].map(embarked_conversion)
```

# Step 4  Machine Learning 

# 1. Logistic Regression 

```python
# import logistic model from sklearn 
from sklearn.linear_model import LogisticRegression

# create instance of the model 
lreg = LogisticRegression()

# training the model 
lreg.fit(x_train , y_train)
```

```python
# testing the model with test data 
y_pred_lreg = lreg.predict(x_test)
print(y_pred_lreg)
```

# Model Evalvation 

```python
# Score it "accuracy_score"
from sklearn.metrics import classification_report , accuracy_score 
print("Classification Model")
print("*--*" * 20 )
lreg_accuracy = round( accuracy_score(y_test , y_pred_lreg)*100  , 2)
print("accuract_score:-" , lreg_accuracy)
print("*--*"*30)

# precsion Recall and Confusion matrix
from sklearn.metrics import recall_score , precision_score , confusion_matrix 
# Precision_score calculation 
precision = precision_score(y_test , y_pred_lreg)
print("Precision_score :-" ,precision)
print("*--*"* 30 )



# Recall_sore calculation 
recall = recall_score(y_test , y_pred_lreg)
print("recall_score:-" , recall)
print("*--*"*30)



# confusion_matrix calculation 
confusion_matrix = confusion_matrix(y_test , y_pred_lreg)
print("confusion_matrix:-" , confusion_matrix)


sns.heatmap(confusion_matrix , annot = True)
plt.title("confusion_matrix for LinearRegression")
plt.show()
```

#  USING deep learning 

```python
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Standardize the features (scaling)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Build a feedforward neural network
model = keras.Sequential([
    layers.Input(shape=(x_train.shape[1],)),  # Input layer
    layers.Dense(128, activation='relu'),     # Hidden layer with 128 neurons and ReLU activation
    layers.Dense(64, activation='relu'),      # Hidden layer with 64 neurons and ReLU activation
    layers.Dense(1, activation='sigmoid')     # Output layer with 1 neuron and sigmoid activation
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=20, batch_size=32, validation_data=(x_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test loss: {loss:.4f}, Test accuracy: {accuracy:.4f}")
```
